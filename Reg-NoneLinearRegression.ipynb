```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Although linear regression can do a great job at modeling some datasets, it cannot be used for all datasets. First recall how linear regression, models a dataset. It models the linear relationship between a dependent variable y and the independent variables x. It has a simple equation, of degree 1, for example y = $2x$ + 3.


Хотя линейная регрессия может отлично справляться с моделированием некоторых наборов данных, ее нельзя использовать для всех наборов данных. Сначала вспомним, как линейная регрессия моделирует набор данных. Он моделирует линейную зависимость между зависимой переменной y и независимыми переменными x. Он имеет простое уравнение степени 1, например y = $ 2x $ + 3.

```python
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph
y = 2*(x) + 3
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
#plt.figure(figsize=(8,6))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

Non-linear regression is a method to model the non-linear relationship between the independent variables $x$ and the dependent variable $y$. Essentially any relationship that is not linear can be termed as non-linear, and is usually represented by the polynomial of $k$ degrees (maximum power of $x$).  For example:

$$ \ y = a x^3 + b x^2 + c x + d \ $$

Non-linear functions can have elements like exponentials, logarithms, fractions, and so on. For example: $$ y = \log(x)$$

We can have a function that's even more complicated such as :
$$ y = \log(a x^3 + b x^2 + c x + d)$$


Нелинейная регрессия - это метод моделирования нелинейной взаимосвязи между независимыми переменными $x$ и зависимой переменной $y$. По существу, любое отношение, которое не является линейным, может быть названо нелинейным и обычно представлено полиномом $ k$ степеней (максимальная степень $ x $).  Например:

$$ \ y = a x^3 + b x^2 + c x + d \ $$

Нелинейные функции могут содержать такие элементы, как экспоненты, логарифмы, дроби и так далее. Например: $$ y = \log(x)$$

У нас может быть функция, которая еще более сложна, например :
$$ y = \log(ax^3 + bx ^2 + cx + d)$$

Let's take a look at a cubic function's graph.

Давайте взглянем на график кубической функции.

```python
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph
y = 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

As you can see, this function has $x^3$ and $x^2$ as independent variables. Also, the graphic of this function is not a straight line over the 2D plane. So this is a non-linear function.


Как вы можете видеть, эта функция имеет $ x ^ 3 $ и $ x ^ 2$ в качестве независимых переменных. Кроме того, график этой функции не является прямой линией на 2D-плоскости. Итак, это нелинейная функция.

Some other types of non-linear functions are:

Некоторыми другими типами нелинейных функций являются:

### Quadratic. Квадратичный


$$ Y = X^2 $$


```python
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

### Exponential

### Экспоненциальный

An exponential function with base c is defined by $$ Y = a + b c^X$$ where b ≠0, c > 0 , c ≠1, and x is any real number. The base, c, is constant and the exponent, x, is a variable.


Экспоненциальная функция с основанием c определяется через $$ Y = a + b c ^ X $$, где b ≠0, c > 0, c ≠1, а x - любое действительное число. Основание, c, является постоянным, а показатель степени, x, является переменной.

```python
X = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

Y= np.exp(X)

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

### Logarithmic

The response $y$ is a results of applying the logarithmic map from the input $x$ to the output $y$. It is one of the simplest form of **log()**: i.e. $$ y = \log(x)$$

Please consider that instead of $x$, we can use $X$, which can be a polynomial representation of the $x$ values. In general form it would be written as\
\begin{equation}
y = \log(X)
\end{equation}


### Логарифмический

Ответ $y$ является результатом применения логарифмического отображения от входного $x$ к выходному $y$. Это одна из простейших форм **log()**: т.е. $$ y = \log(x)$$

Пожалуйста, учтите, что вместо $x$ мы можем использовать $X$, который может быть полиномиальным представлением значений $x$. В общем виде это было бы записано как\
\begin{equation}
y = \log(X)
\end{equation}

```python
X = np.arange(-5.0, 5.0, 0.1)

Y = np.log(X)

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

### Sigmoidal/Logistic

### Сигмоидальный/Логистический

$$ Y = a + \frac{b}{1+ c^{(X-d)}}$$


```python
X = np.arange(-5.0, 5.0, 0.1)


Y = 1-4/(1+np.power(3, X-2))

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

# Пример нелинейной регрессии

For an example, we're going to try and fit a non-linear model to the datapoints corresponding to China's GDP from 1960 to 2014. We download a dataset with two columns, the first, a year between 1960 and 2014, the second, China's corresponding annual gross domestic income in US dollars for that year.

Для примера мы попытаемся подогнать нелинейную модель к точкам данных, соответствующим ВВП Китая с 1960 по 2014 год. Мы загружаем набор данных с двумя столбцами: первый - год между 1960 и 2014 годами, второй - соответствующий годовой валовой внутренний доход Китая в долларах США за этот год.

### Plotting the Dataset. Построение графика набора данных

This is what the datapoints look like. It kind of looks like an either logistic or exponential function. The growth starts off slow, then from 2005 on forward, the growth is very significant. And finally, it decelerates slightly in the 2010s.

Вот как выглядят точки данных. Это похоже либо на логистическую, либо на экспоненциальную функцию. Рост начинается медленно, затем, начиная с 2005 года, рост становится очень значительным. И, наконец, в 2010-х годах он немного замедляется.

```python
plt.figure(figsize=(8,5))
x_data, y_data = (df["Year"].values, df["Value"].values)
plt.plot(x_data, y_data, 'ro')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

### Choosing a model. Выбор модели

From an initial look at the plot, we determine that the logistic function could be a good approximation,
since it has the property of starting with a slow growth, increasing growth in the middle, and then decreasing again at the end; as illustrated below.

При первоначальном взгляде на график мы определяем, что логистическая функция может быть хорошим приближением,
поскольку она обладает свойством начинать с медленного роста, увеличивать рост в середине, а затем снова уменьшаться в конце; как показано ниже:

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = 1.0 / (1.0 + np.exp(-X))

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

The formula for the logistic function is the following. Формула для логистической функции выглядит следующим образом:

$$ \hat{Y} = \frac1{1+e^{-\beta\_1(X-\beta\_2)}}$$

$\beta\_1$: Controls the curve's steepness (Контролирует крутизну кривой),

$\beta\_2$: Slides the curve on the x-axis (Перемещает кривую по оси x).


### Building The Model. Построение Модели

Now, let's build our regression model and initialize its parameters.

Теперь давайте построим нашу регрессионную модель и инициализируем ее параметры.

```python
def sigmoid(x, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y
```

Lets look at a sample sigmoid line that might fit with the data:

Давайте посмотрим на образец сигмовидной линии, которая может соответствовать данным:

```python
beta_1 = 0.10
beta_2 = 1990.0

#logistic function
Y_pred = sigmoid(x_data, beta_1 , beta_2)

#plot initial prediction against datapoints
plt.plot(x_data, Y_pred*15000000000000.)
plt.plot(x_data, y_data, 'ro')
```

Our task here is to find the best parameters for our model. Lets first normalize our x and y:

Наша задача здесь - найти наилучшие параметры для нашей модели. Давайте сначала нормализуем наши x и y:

```python
# Lets normalize our data
xdata =x_data/max(x_data)
ydata =y_data/max(y_data)
```

#### How we find the best parameters for our fit line?

we can use **curve_fit** which uses non-linear least squares to fit our sigmoid function, to data. Optimize values for the parameters so that the sum of the squared residuals of sigmoid(xdata, \*popt) - ydata is minimized.

popt are our optimized parameters.


#### Как мы находим наилучшие параметры для нашей линейки fit?

мы можем использовать **curve_fit**, который использует нелинейные наименьшие квадраты для подгонки нашей сигмоидной функции к данным. Оптимизируйте значения параметров таким образом, чтобы сумма квадратов остатков sigmoid(xdata, \*pop) - ydata была сведена к минимуму.

popt - это наши оптимизированные параметры.

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, xdata, ydata)
#print the final parameters
print(" beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))
```

Now we plot our resulting regression model.

Теперь мы построим нашу результирующую регрессионную модель.

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize=(8,5))
y = sigmoid(x, *popt)
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,y, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

## Practice

Can you calculate what is the accuracy of our model?

Можете ли вы рассчитать, какова точность нашей модели?

```python
# write your code here

# split data into train/test
msk = np.random.rand(len(df)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]

# build the model using train set
popt, pcov = curve_fit(sigmoid, train_x, train_y)

# predict using test set
y_hat = sigmoid(test_x, *popt)

# evaluation
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_hat - test_y) ** 2))
from sklearn.metrics import r2_score
print("R2-score: %.2f" % r2_score(test_y,y_hat) )
```
